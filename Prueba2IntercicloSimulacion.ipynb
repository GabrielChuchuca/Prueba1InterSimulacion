```python
class Institucion():
  """cada clase dura 41 minutos y son 8 clases al dia y 30 minutos de receso que dan 6 horas dirais"""
  def __init__(self, env, num_aulas, nombre_institucion):
    self.env = env
    self.aulas = simpy.Resource(env, num_aulas)
```

```python
es_x_aula = round(N_ESTUDIANTES / AULAS)
es_x_aula
```

```python
def llenado_doc(env, docente, colegio, au):
   #d = list(range(docente))
   #a = random.sample(d, 15)
   with colegio.aulas.request() as aula:   
    if docente>0:
      print("El docente {} llega en el tiempo {} al aula {}".format(docente, env.now, au + 1)) if au < 15 else None
      yield env.timeout(0.001)
      print("El docente {} inicia la clase en el tiempo {} al aula {}".format(docente, env.now, au + 1)) if au < 15 else None
      yield aula
```

```python
def simulacion(env, num_aulas, estudiantes, docentes):
  institucion = Institucion(env, num_aulas, "FRANCISCO TAMARIZ")
  print("***LLEGADA DE LOS ALUMNOS A LAS AULAS***")
  for i in range(estudiantes):
        env.process(llenado(
            env,(i+1),institucion))
  
  yield env.timeout(0.002)
  print("***LLEGADA DE LOS DEOCENTES A LAS AULAS***")
  d = list(range(docentes))
  a = random.sample(d, 15)
  for j in range(num_aulas):
    env.process(llenado_doc(env, a[j]+ 1, institucion, j))
  yield env.timeout(0)
 
```

```python
if __name__ == "__main__":
  print("Punto de vacunacion en Cuenca")
  # Creamos el entorno de simulacion
  env=simpy.Environment()
  print("# de Estudinates vacunados o con certificado => ", ESTUDIANTES_VACUNADOS)
  env.process(simulacion(env, AULAS, ESTUDIANTES_VACUNADOS, DOCENTES_VACUNADOS))
  # Ejecutamos el proceso durante el tiempo de simulacion
  env.run(until = 1000)
```

```python
class Car(object):
  def __init__(self, env):
    self.env = env
    # Start the run process everytime an instance is created.
    self.action = env.process(self.run())

  def run(self):
    a = 0
    while a <4:
      print('Start parking and charging at %d' % self.env.now)
      charge_duration = 5
      # We yield the process that process() returns
      # to wait for it to finish
      yield self.env.process(self.charge(charge_duration))

      # The charge process has finished and
      # we can start driving again.
      print('Start driving at %d' % self.env.now)
      trip_duration = 2
      yield self.env.timeout(trip_duration)
      a = a + 1

  def charge(self, duration):
    yield self.env.timeout(duration)
```
